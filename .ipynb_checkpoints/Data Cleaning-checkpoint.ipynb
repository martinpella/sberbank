Training: there are 32 observations with full_sq < life_sq.

```python
train_df.loc[(train_df['full_sq'] < train_df['life_sq']) & (train_df['life_sq'] - train_df['full_sq'] < 5), :]
```

Training: there are 13 observations with full_sq < life_sq and the difference between them is very small (less than 5 sq meters). In these cases maybe the column values are inverted.

```python
temp_full_sq = train_df.loc[(train_df['full_sq'] < train_df['life_sq']) & (train_df['life_sq'] - train_df['full_sq'] < 5), 'full_sq']
temp_life_sq = train_df.loc[(train_df['full_sq'] < train_df['life_sq']) & (train_df['life_sq'] - train_df['full_sq'] < 5), 'life_sq']
bad_ix = train_df.loc[(train_df['full_sq'] < train_df['life_sq']) & (train_df['life_sq'] - train_df['full_sq'] < 5), :].index
```

```python
train_df.ix[bad_ix, 'full_sq'] = temp_life_sq
train_df.ix[bad_ix, 'life_sq'] = temp_full_sq
```

```python
train_df.loc[(train_df['full_sq'] < train_df['life_sq']) & (train_df['life_sq'] > 150), :]
```

Training: there are 13 observations with full_sq < life_sq and also life_sq is a very large value, much more than the 75% percentile.
In this cases I consider full_sq as correct value and set life_sq to NaN

```python
bad_ix = train_df.loc[(train_df['full_sq'] < train_df['life_sq']) & (train_df['life_sq'] > 150), :].index
train_df.ix[bad_ix, 'life_sq'] = np.NaN
```

Training: for the remaining 6 observations with full_sq < life_sq, set full_sq as NaN

```python
bad_ix = train_df.loc[train_df['full_sq'] < train_df['life_sq'], :].index
train_df.ix[bad_ix, 'full_sq'] = np.NaN
```
