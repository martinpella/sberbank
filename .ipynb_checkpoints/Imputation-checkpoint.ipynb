```python
def null_qty(data):
    df_null = data.isnull().sum().sort_values(ascending=False)
    df_null = df_null[df_null > 0]
    return pd.DataFrame(data = {'Qty': df_null.values,
                                '%': (df_null / data.shape[0]).apply(lambda x: round(x, 3)).values},
                        columns = ['Qty', '%'],
                        index = df_null.index)
```

```python
null_train = null_qty(train_df)
```

```python
null_test = null_qty(test_df)
```
